```python
#importing libraries
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
#delimiter is to specify that we are using tab seperated file
#quoting(important when working with text)
#quoting=3 means ignore all the double quotes
data_set=pd.read_csv("new_file.csv")
data_set
del data_set["Unnamed: 0.1"]
del data_set["Unnamed: 0"]
data_set
```

```python
import re
import nltk
#stopwords- words which are not relevant for predicting the output.. like "a,the,an etc."
nltk.download("stopwords")
#importing all the stopwords we just downloaded
from nltk.corpus import stopwords
#stemming consist of taking only root of word.. that indicate enough what does it mean...
from nltk.stem.porter import PorterStemmer
#corpus will contain different reviews but all cleaned reviews
corpus = []
for i in range(0, 500):
    #we replace any element which is not a alphabet is replaced by " ", we replace puntuation
    #we use "^" means we dont want to replace upper and lower case character
    review = re.sub('[^a-zA-Z]', ' ', data_set['Review'][i])
    #we convert all uppercase letter in lower case
    review = review.lower()
    #we are gonna split review is different word so we can apply stemming to different words
    review = review.split()
    ps = PorterStemmer()
    #list of all the word which are not used for predicting data
    all_stopwords = stopwords.words('english')
    all_stopwords.remove('not')
    all_stopwords.remove("isn't")
    review = [ps.stem(word) for word in review if not word in set(all_stopwords)]
    #again joining the review to get back in original form
    review = ' '.join(review)
    corpus.append(review)
```

```python
print(corpus)
```

```python
#creating a bag of words model
from sklearn.feature_extraction.text import CountVectorizer
#even after data cleaning some word doesnt make sense like "texture", hence we limit the number of features..
#we choose the most frequent words
cv=CountVectorizer(max_features=999)
x = cv.fit_transform(corpus).toarray()
y = data_set.iloc[:, -1].values
```

```python
#splitting the data in test set and train set
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.20, random_state = 0)
```

```python
#train using naive biased model
from sklearn.naive_bayes import GaussianNB
classifier = GaussianNB()
classifier.fit(x_train, y_train)
```

```python
y_pred = classifier.predict(x_test)
print(np.concatenate((y_pred.reshape(len(y_pred),1), y_test.reshape(len(y_test),1)),1))
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score
cm = confusion_matrix(y_test, y_pred)
print(cm)
accuracy_score(y_test, y_pred)
```
