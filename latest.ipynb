```python
import pandas as pd
import numpy as np
!pip install scikit-multilearn
from skmultilearn.adapt import MLkNN
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics import accuracy_score
```

```python
aspects_df = table.copy()
aspects_df.head()
aspects_df.iloc[0]
aspects_df.head()
```

```python
X=table_rev["review1"]
y = np.asarray(aspects_df[aspects_df.columns[1:]])
X
#y
```

```python
vetorizar = TfidfVectorizer(max_features=100,max_df=0.85)
vetorizar.fit(X)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=0)
#print((y_train))
for i in y_train:
    if len(set(i))==1:
        print("YES")
```

```python
X_train_tfidf = vetorizar.transform(X_train)
X_test_tfidf = vetorizar.transform(X_test)
```

```python
X_train_tfidf.shape,y_train.shape
```

```python
mlknn_classifier = MLkNN()
mlknn_classifier.fit(X_train_tfidf, y_train)
score = mlknn_classifier.score(X_test_tfidf, y_test)
print(y_train)
print((y_test))
print(score)
```

```python
# using classifier chains
from skmultilearn.problem_transform import ClassifierChain
from sklearn.linear_model import LogisticRegression
# initialize classifier chains multi-label classifier
classifier = ClassifierChain(LogisticRegression())
# Training logistic regression model on train data
classifier.fit(X_train_tfidf, y_train)
# predict
predictions = classifier.predict(X_test_tfidf)
# accuracy
print("Accuracy = ",accuracy_score(y_test,predictions))
print("\n")
```

```python
from skmultilearn.adapt import MLkNN
from scipy.sparse import csr_matrix, lil_matrix
classifier_new = MLkNN(k=10)
# Note that this classifier can throw up errors when handling sparse matrices.
x_train = lil_matrix(X_train_tfidf).toarray()
y_train = lil_matrix(y_train).toarray()
x_test = lil_matrix(X_test_tfidf).toarray()
# train
classifier_new.fit(x_train, y_train)
# predict
predictions_new = classifier_new.predict(x_test)
# accuracy
print("Accuracy = ",accuracy_score(y_test,predictions_new))
print("\n")
```

```python
#CAMERA RATING
from textblob import TextBlob

pol = lambda x: 5*(TextBlob(x).sentiment.polarity)
sub = lambda x: TextBlob(x).sentiment.subjectivity

camera_df['polarity'] = camera_df['Review'].apply(pol)
camera_df['subjectivity'] = camera_df['Review'].apply(sub)
camera_df
```

```python
#Battery RATING
from textblob import TextBlob

pol = lambda x: 5*(TextBlob(x).sentiment.polarity)
sub = lambda x: TextBlob(x).sentiment.subjectivity

battery_df['polarity'] = battery_df['Review'].apply(pol)
battery_df['subjectivity'] = battery_df['Review'].apply(sub)
battery_df
```

```python
#PRICE RATING

from textblob import TextBlob

pol = lambda x: 5*(TextBlob(x).sentiment.polarity)
sub = lambda x: TextBlob(x).sentiment.subjectivity

price_df['polarity'] = price_df['Review'].apply(pol)
price_df['subjectivity'] = price_df['Review'].apply(sub)
price_df
```

```python
#DISPLAY RATING

from textblob import TextBlob

pol = lambda x: 5*(TextBlob(x).sentiment.polarity)
sub = lambda x: TextBlob(x).sentiment.subjectivity

display_df['polarity'] = display_df['Review'].apply(pol)
display_df['subjectivity'] = display_df['Review'].apply(sub)
display_df
```

```python
cam_avg = camera_df.mean(axis=0)
bat_avg = battery_df.mean(axis=0)
pri_avg = price_df.mean(axis=0)
dis_avg = display_df.mean(axis=0)
print(cam_avg)
print(bat_avg)
print(pri_avg )
print(dis_avg)
```

```python
import matplotlib.pyplot as plt
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
features = ['Camera', 'Battery', 'Price', 'Display']
ratings = [cam_avg.polarity , bat_avg.polarity , pri_avg.polarity , dis_avg.polarity]
plt.xlabel("FEATURES")
plt.ylabel("RATING")
ax.bar(features,ratings)
plt.show()
```
