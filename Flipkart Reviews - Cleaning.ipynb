```python
# Apply a first round of text cleaning techniques
import re
import string

def clean_text_round1(text):
    '''Make text lowercase, remove punctuation and remove words containing numbers,remove'READ MORE',remove emojis.'''
    text = text.lower()
    #text = re.sub('\[.*?\]', '', text)
    text = re.sub('[%s]' % re.escape(string.punctuation), '', text)
    text = re.sub('\w*\d\w*', '', text)
    #text = re.sub('\w*\\w*', '', text)
    text = re.sub(r"read more", "", text, flags=re.I)
    RE_EMOJI = re.compile('[\U00010000-\U0010ffff]', flags=re.UNICODE)
    text = RE_EMOJI.sub(r'', text)
    text = re.sub('[^a-z] ', '', text)
    #text = re.sub(r'\W',"",text,flags=re.I)j
    #text = str.replace(/[^a-z0-9-]/g, '');
    return text
round1 = lambda x: clean_text_round1(x)
```

```python
# Let's take a look at the updated text
corpus['Review'] = pd.DataFrame(corpus.Review.apply(round1))
corpus['Heading'] = pd.DataFrame(corpus.Heading.apply(round1))
corpus
```

# Applying porter stemming here as many words are mixed at some places (that is types without space ,so it they wont make sense during sentiment analysis.  Therefore i am thinking of using porter stemming for now and not lemmatization. Later we an figure out if lancester is required)

STEMMING : cuts out the styrating or ending part of a word : types --> 
        1)porter pst(less agressive) eg : given -> give
        2)lancester lst(more agressive) eg: givem -> giv
Lemmatization : converts into proper word


```python
#ADDING A COLUMN OF TOKENS OF THE REVIEWS
from nltk.tokenize import sent_tokenize, word_tokenize
corpus['tokens'] = corpus['Review'].apply(lambda x: word_tokenize(x))
corpus
```

```python
from wordsegment import load,segment
load()
def word_seg(text):
    stem_text =[]
    for i in text:
        stem_text=stem_text+segment(i)
    return stem_text
corpus['tokens_seg'] = corpus['tokens'].apply(lambda x: word_seg(x))
corpus
```

```python
import re
import nltk
from nltk.corpus import stopwords
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
all_stopwords = stopwords.words('english')
all_stopwords.remove('not')
all_stopwords.remove("isn't")
def portstem(text):
    review = [ps.stem(word) for word in text if not word in set(all_stopwords)]
    return review
corpus['tokens_stem'] = corpus['tokens_seg'].apply(lambda x: portstem(x))
corpus
```

```python
a=[]
for i in range(500):
    a.append(" ".join(corpus["tokens_stem"][i]))
corpus["review1"]=a
corpus.to_csv("flip.csv")
```

```python
from textblob import TextBlob
pol = lambda x:5*TextBlob(" ".join(x)).sentiment.polarity
sub = lambda x:TextBlob(" ".join(x)).sentiment.subjectivity
corpus['polarity']=corpus['tokens_stem'].apply(pol)
corpus['subjectivity']=corpus['tokens_stem'].apply(sub)
corpus['polarity1']=corpus['tokens_seg'].apply(pol)
corpus['subjectivity1']=corpus['tokens_seg'].apply(sub)
```
